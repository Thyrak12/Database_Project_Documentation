#### Create table and insert into User

```python
import random
import string
import mysql.connector  # Replace with your database connector library (e.g., psycopg2 for PostgreSQL)
from faker import Faker
from datetime import datetime

faker = Faker();

# Database connection details (replace with your credentials)
DB_HOST = "localhost"
DB_USER = "root"
DB_PASSWORD = ""
DB_NAME = "test_database"  # Replace with your database name

# Function to generate random data (you can customize this)
def generate_user_data():
  roles = ['admin', 'staff', 'customer']
  name = faker.name()
  email = faker.email()
  password = faker.password()
  role = random.choice(roles)
  created_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
  return name, email, password, role, created_at

# Connect to the database
try:
    connection = mysql.connector.connect(host=DB_HOST, user=DB_USER, password=DB_PASSWORD, database=DB_NAME)
    cursor = connection.cursor()
except mysql.connector.Error as err:
  print("Error connecting to database:", err)
  exit()

# Create table from file ---------------------------------------

# Open and read the file as a single buffer
fd = open('../Database_Implementation/Project_script.sql', 'r')
sqlFile = fd.read()
fd.close()

# all SQL commands (split on ';')
sqlCommands = sqlFile.split(';')

# Execute every command from the input file
for command in sqlCommands:
  if command.__len__() > 10: # Filter out the empty string
    cursor.execute(command)
connection.commit()

# Done creating tables -----------------------------------------

# Generate and insert data
print('Start inserting ...')
insertion_amount = 10
num_records = 1000000
i = 0
for k in range(insertion_amount):
  for _ in range(num_records):
    name, email, password, role, created_at = generate_user_data()
    insert_query = """
    INSERT INTO User (name, email, password, role, created_at)
    VALUES (%s, %s, %s, %s, %s)
    """
    cursor.execute(insert_query, (name, email, password, role, created_at))
  connection.commit()
  i = i + 1
  print('Insertion round', i, num_records, 'records.')

print(f"Successfully inserted {num_records*insertion_amount} records into the database.")

# Close the connection
cursor.close()
connection.close()

print('Finished inserting!!!')
```

#### Generate stock

```python
import random
import mysql.connector
from datetime import datetime

# Database connection details (replace with your credentials)
DB_HOST = "localhost"
DB_USER = "root"
DB_PASSWORD = ""
DB_NAME = "test_database"

# Function to generate random data (you can customize this)
def generate_stock_data(var):
  variant_id = var
  qty = random.randint(1, 1000)
  type = random.choice(['in', 'out'])
  created_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
  return variant_id, qty, type, created_at

# Connect to the database
try:
    connection = mysql.connector.connect(host=DB_HOST, user=DB_USER, password=DB_PASSWORD, database=DB_NAME)
    cursor = connection.cursor()
except mysql.connector.Error as err:
  print("Error connecting to database:", err)
  exit()

# Generate and insert data
MAX_STOCK_PER_VARIANT = 3

cursor.execute('SELECT id FROM Product_Variant ORDER BY id ASC')
fetched_ids = cursor.fetchall()
variant_ids = [row[0] for row in fetched_ids]
total_variant = len(variant_ids)
print('Amount of product variant: ', len(fetched_ids))

if not variant_ids:
  print("No variant IDs found or fetched. Cannot generate stocks. Exiting.")
  exit

print('Start inserting ...')
i = 0
total_record = 0
for k in range(total_variant):
  stocking = random.randint(1, MAX_STOCK_PER_VARIANT)
  for _ in range(stocking):
    variant_id, qty, type, created_at = generate_stock_data(variant_ids[k])
    insert_query = """
    INSERT INTO Stock_Transaction (variant_id, qty, type, created_at)
    VALUES (%s, %s, %s, %s)
    """
    cursor.execute(insert_query, (variant_id, qty, type, created_at))
  connection.commit()
  i = i + 1
  print('Insertion variant', i, stocking, 'records.')
  total_record += stocking

print(f"Successfully inserted {total_record} records into the database.")

# Close the connection
cursor.close()
connection.close()

print('Finished inserting!!!')
```
